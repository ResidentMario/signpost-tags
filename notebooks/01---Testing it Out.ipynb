```python
import mwapi
```

```python
session = mwapi.Session('https://en.wikipedia.org', user_agent='test-notebook')
```

```python
example_output = session.get(action='parse', format='json', page='Wikipedia:Wikipedia Signpost/2016-10-14/News and notes', prop='text')
```

```python
example_text = example_output['parse']['text']['*']
```

```python
import bs4

soup = bs4.BeautifulSoup(example_text, 'html.parser')
```

```python
fulltext = " ".join(list(soup.stripped_strings))
fulltext_end_token = fulltext.rfind('Previous "News and notes"')
fulltext_start_token = fulltext.find('News and notes')
parsed_text = fulltext[fulltext_start_token + len('News and notes') + 1:fulltext_end_token - 3]
parsed_text = parsed_text.replace("Share this E-mail Facebook Twitter Google+ LinkedIn Reddit StumbleUpon Delicious Digg", "")
```
